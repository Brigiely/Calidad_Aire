## **Evaluación de la Calidad del Aire y las Condiciones Climáticas en el Perú**  


---
## **1. Introduccion**  
Este informe presenta un análisis comparativo entre calidad del aire y las condiciones climáticas en diversas ciudades del Perú, utilizando datos extraídos en tiempo real de las APIs **IQAir** y **WeatherAPI**.

Este proyecto tuvo como objetivo identificar las ciudades más adecuadas para la salud respiratoria a partir de parámetros como el Índice de Calidad del Aire (AQI), temperatura y humedad. Se implementaron técnicas de procesamiento de datos, análisis y visualizaciones efectivas para facilitar la interpretación de los resultados.

---
## **2. Objetivos**

### **Objetivo General**  
Evaluar las condiciones climáticas y la calidad del aire en distintas ciudades del Perú para identificar las zonas más adecuadas para vivir, priorizando la salud respiratoria.

### **Objetivos Específicos**  
- Extraer datos sobre la calidad del aire (AQI) y condiciones climáticas (temperatura, humedad y viento) mediante APIs confiables y de acceso público.  
- Analizar y comparar los datos obtenidos para identificar patrones relevantes y correlaciones entre calidad del aire y condiciones climáticas.   
- Clasificar las ciudades en función de los riesgos asociados a la calidad del aire para salud respiratoria, para poder escoger una la mejor opcion
  entre todas las ciudades para vivir.  
- Proporcionar recomendaciones basadas en los datos obtenidos para la mejora de la calidad de vida en las ciudades con condiciones menos favorables.  


## **3. Planeamiento de Extracción de Datos**

## **4. Metodología**  

1. **Planificación y Selección de Fuentes de Datos**  
   - Las APIs fueron seleccionadas por su disponibilidad, fiabilidad y alcance geográfico, permitiendo obtener información actualizada de las ciudades peruanas.  
   - Se revisaron las instrucciones de uso de cada API para entender cómo usarlas correctamente, y se procedio a acceder a los datos que fueron necesarios para lograr los objetivos.

2. **Extracción de Datos**  
   - Se configuraron dos funciones específicas en Python para interactuar con cada API:  
     - `obtener_clima_weatherapi`: Recupera datos climáticos de WeatherAPI.  
     - `obtener_calidad_aire_iqair`: Recupera datos de calidad del aire de IQAir.  
  

3. **Procesamiento de los Datos Extraídos**
   - Los datos se procesan extrayendo información relevante de las APIs seleccionadas: WeatherAPI y IQAir.
     
         -Temperatura (°C) y humedad (%) desde WeatherAPI.
         -Índice de Calidad del Aire (AQI) desde IQAir, utilizando la latitud y longitud obtenida previamente.
     
   - Los datos fueron descargados desde las APIs en formato JSON y luego transformados en listas especificas para cada variable (temperatura, humedad y AQI).
 

5. **Organización y Manejo de Datos**  
   - Los datos de las dos APIs se almacenaron en listas separadas para cada variable. Esta organizacion permitio manejar los datos de manera eficiente para su analisis y visualizacion sin necesidad de transformaciones intermedias. 

6. **Validación y Organización de los Datos**  
   - Se implementaron verificaciones para manejar posibles errores en la respuesta de las APis, como datos faltantes o respustas invalidas.
   - Cuando una API no devuelve datos validos, se asignan valores predeterminados (0) para garantizar la continuidad del analisis.

1. **Arequipa** es la ciudad más habitable del análisis, con un balance óptimo entre calidad del aire, temperatura y humedad.

2. **Para Autoridades:**
   - Priorizar proyectos de mitigación en ciudades con alto riesgo, como campañas de reducción de emisiones en **Lima**.

3. **Para Viajeros:**
   - Seleccionar destinos como **Arequipa** o **Piura** para una estancia saludable.
